## Экспоненциальное распределение
**Экспоненциальное распределение** - непрерывное распределение, моделирующее время между двумя последовательными свершениями одного и того же события.
**Функция распределения:**
$$ F_X(x)=  \left\{
\begin{array}{ll}
      1 - e^{-{\lambda}x} & ,x \geq 0 \\
      0 & ,x<0 \\
\end{array}
\right. $$
**Плотность вероятности:**
$$ f_X(x) = \left\{
\begin{array}{ll}
      \lambda e^{-{\lambda}x} & ,x \geq 0 \\
      0 & ,x<0 \\
\end{array}
\right. $$
**Математическое ожидание:**
$$ \mathbb{E}(X) = {\lambda}^{-1}$$
**Дисперсия:**
$$ \mathbf{D}(x) = \frac{1}{\lambda^{2}} $$


Для своей задании, я использую экспоненциальное распределение с $\lambda = 0.2$

### Генерация генеральной выборки из 1000 значений и его гисторгамм, и график теоретической функции распределения

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
l = 0.1 #lambda
E = 1 / l
D = 1 / (l ** 2)

general_sample = np.random.exponential(size=1000, scale=1/l)
plt.hist(general_sample, density=True, bins=50)
plt.plot(np.linspace(0, 50, 1000), sts.expon(scale = 1/l).pdf(np.linspace(0, 50, 1000)))
plt.ylabel('PDF')
plt.xlabel('$x$')
```

### Распределение средней выборки для выборок размера 5
$$\mu = \mathbb{E} $$
$$\sigma = \sqrt{D / 5} = 2.236$$

```python
means_5 = []
for _ in range(1000):
    means_5.append(np.mean(np.random.exponential(scale=1/l, size=5)))
plt.hist(means_5, bins=50, density=True)
plt.plot(np.linspace(0, 30, 1000), sts.norm(loc=E, scale=((D / 5) ** 0.5)).pdf(np.linspace(0, 30, 1000)))
plt.ylabel('PDF')
plt.xlabel('$x$')
```

### Распределение средней выборки для выборок размера 10
$$\mu = \mathbb{E} $$
$$\sigma = \sqrt{D / 10} = 1.581$$

```python
means_10 = []
for _ in range(1000):
    means_10.append(np.mean(np.random.exponential(scale=1/l, size=10)))
plt.hist(means_10, bins=50, density=True)
plt.plot(np.linspace(0, 30, 1000), sts.norm(loc=E, scale=((D / 10) ** 0.5)).pdf(np.linspace(0, 30, 1000)))
plt.ylabel('PDF')
plt.xlabel('$x$')
```
